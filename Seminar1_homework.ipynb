### Задание 1 .
**Выберите любой продукт.  
Например ваш продукт  
ozon.ru  
citilink.ru  
telegram  
Какие бы ключевые метрики вы выделили? Опишите почему вы выбрали именно их и можно ли их измерить.**




### Решение.
Для анализа выбрал сайт Яндек.Учебник (education.yandex.ru).  
Выделим несколько ключевых метрик для данного продукта:  
* Места на странице, куда чаще всего кликают пользователи.
* Что из предложенного на странице видит большинство пользователей, заставляет ли предложенная на странице информация, прокручивать сайт ниже.
* Траектории путей движения пользователей по сайту и откуда приходят пользователи.
* Какие продукты пользуются большей популярностью - для продвинутых учеников или для отстающих учеников, пытающихся восполнить пробелы в знаниях.

Данные метрики я выбрал потому, что качество сайта (тем более с образовательными элементами) во многом зависит от его удобства в использовании, т.е "юзабилити". Чем удобнее сайт, тем быстрее и проще посетитель сможет достичь своей цели. Все предложенные метрики могут быть измерены, а для первых трех в Яндекс.Метрике есть инструменты, дающие статистически значимые данные для анализа поведения пользователей на сайте.


### Задание 2.
**Сформулируйте 5-10 гипотез для выбранного продукта.**

### Решение.

#### Гипотезы:
1. Если слева создать дополнительное меню с наиболее популярными разделами, количество прокруток страницы снизится. Пользователям будет удобнее находить необходимую информацию.
2. Размещение полезной информации в виде небольших баннеров о различных возможностях сервиса, а также вывод популярных разделов позволит сделать главную страницу более полезной для пользователей и продвижения.
3. Я считаю, что идеальные пользователи Яндекс.Учебника - это ученики школ, их родители и учителя, которые живут в небольших городах и сельской местности, где доступ к образовательным ресурсам ограничен.
4. Предоставить возможность регистрации ученика по электронной почте и через профили социальных сетей, т.к. существует лишь один способ входа на сайт с использование данных, которые заранее должен ввести учитель.
5. Мне кажется, что если добавить короткие обучающие видео (как в тиктоке), то вероятность то, что пользователь заинтересуется какими-то темами увеличится.
