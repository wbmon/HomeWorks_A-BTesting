### Задание1.
На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem.
Распишите, какие метрики и как вы будете анализировать.

#### Решение:
Для анализа на первом семинаре выбрал сайт Яндек.Учебник (education.yandex.ru).  
Составим план проведения A/B тестирования.  
1. Формулируем гипотезу: Если слева создать дополнительное меню с наиболее популярными разделами, количество прокруток страницы снизится. Пользователям будет удобнее находить необходимую информацию.
2. Определяемся с целевым показателем — что в итоге будет критерием: количество прокруток страницы.
3. Выбераем один элемент тестирования. Добавленное слева меню.
4. Определяем выборку. Для этой цели можно использовать калькулятор.
5. Определяем продолжительность A/B тестирования. В Google Experiments минимальное время тестирования по умолчанию — 14 дней.
6. Проверяем однородность аудитории. После запуска теста поток посетителей будет распределяться 50/50 и важно, чтобы эти две части были максимально однородными и похожими. То есть на конечный результат должен влиять только эксперимент. Чтобы проверить однородность, нужно сначала провести A/A тестирование — поделить трафик, но показывать абсолютно одинаковые страницы.
7. Принимаем во внимание внешние факторы. Сайт Яндекс.Учебник может быть по-разному востребован в период праздничных дней, школьных каникул, во время подготовки к экзаменам и т.д.
8. Не останавливаем эксперимент, пока не пройдет время, которое определено ранее и пока не наберется достаточное количество человек. «Просматривать» данные — это нормально, главное - удержаться от остановки эксперимента до того, как он завершится.
9. Повторное тестирование. По окончании теста проверяем статистическую значимость результатов. Но даже если данные достоверны, эксперимент продолжался достаточное количество времени и один из вариантов уверенно победил, провести тестирование еще раз — обычная и хорошая практика.
